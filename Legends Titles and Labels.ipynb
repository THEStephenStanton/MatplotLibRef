```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [4,2,7]

x2 = [1,2,3]
y2 = [10, 14, 12]

plt.plot(x, y, label="First line")
plt.plot(x2, y2, label="Second line")

plt.legend()

plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Awesome Title\nCheck It Out")

plt.show()
```
