```python
import matplotlib.pyplot as plt
```

```python
days = [1,2,3,4,5,6,7]

sleeping = [5,7,8,7,5,6,9]
eating =   [2,5,4,1,8,4,3]
working =  [7,8,9,2,3,10,7]
playing =  [10,4,3,14,8,4,5]

plt.stackplot(days, sleeping, eating, working, playing, colors=["m", "c", "r", "k"])

# Labeling
plt.plot([], [], c="k", label="Playing", linewidth=3)
plt.plot([], [], c="r", label="Working", linewidth=3)
plt.plot([], [], c="c", label="Eating", linewidth=3)
plt.plot([], [], c="m", label="Sleeping", linewidth=3)


plt.legend()

plt.xlabel("Day")
plt.ylabel("Hours Spent")
plt.show()
```
